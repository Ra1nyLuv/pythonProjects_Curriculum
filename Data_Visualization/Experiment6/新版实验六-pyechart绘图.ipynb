```python
#Graph
import pandas as pd
from pyecharts import options as opts 
from pyecharts.charts import Graph
r1=pd.read_csv(r"t_alibaba_data3.txt",names=["user","brand","behavr","date"],sep="\t",dtype={"behavr":int})
#pandas会自己判断数据类型，但是有时也需要自己额外指定数据类型
r1.date="2011/"+r1.date
r1.date=pd.to_datetime(r1.date)

links=[] 
user=set([]) 
brand=set([])
for i in r1[(r1.date>"2011-07-04")&(r1.date<"2011-07-10")].groupby(["brand","user"]).size().iteritems():
    links.append({"source":str(i[0][0]),"target":str(i[0][1])}) 
    user.add(str(i[0][0])) 
    brand.add(str(i[0][1]))
nodes=[] 
for i in user:
    nodes.append( {"name":i, "category":0}) 
for i in brand:
    nodes.append({"name":i, "category":1})
G1=Graph()
G1.add("",nodes, links, [ {"name":"user"},{"name":"item"}])
G1.set_series_opts(label_opts=opts.LabelOpts(is_show=False)) 
G1.render_notebook()
```

```python
#平行坐标系
import pyecharts 
import numpy as np#的建店调的特征柜萨
a=pd.read_csv(r"shoes.csv")
a.sales=a.sales.astype(str).str.split("人",expand=True)[0] 
a.sales=a.sales.astype(np.int64)
a.price =a.price.astype(float)
z1=a.sales*a.price 
z1.name="xse"
a1=pd.concat([a,z1],axis=1)
a1
```

```python

te_zheng={"nick":[],"z_xse":[],"z_num":[],"p_sales":[],"p_bdj":[],"p_price":[]}
for i in a1.groupby("nick"):
    te_zheng["nick"]. append(i[0]) 
    te_zheng["z_xse"]. append(i[1].xse.sum() )
    te_zheng["z_num"].append(len(i[1]))
    te_zheng["p_sales"].append(round(i[1].sales.mean(),1)) 
    if i[1].sales.sum()==0:
        te_zheng["p_bdj"].append(0) 
    else:
        te_zheng["p_bdj"].append(round(i[1].xse.sum()/i[1].sales.sum(),1)) 
    te_zheng["p_price"].append(round(i[1].price.mean(),1))
df_te_zheng=pd.DataFrame(te_zheng)
df_te_zheng.sort_values(by="z_xse", ascending=False, inplace=True)
df_te_zheng.head()
```

```python
#平行坐标系
from pyecharts import options as opts 
from pyecharts.charts import Page,Parallel
z1=df_te_zheng.head(5)
for i in z1.iterrows():
    print(i[1].tolist()[1:]) 
data=[]
#按教提保存在小心中
for i in z1.iterrows():
    data.append(i[1].tolist()[1:])
c=Parallel().add_schema(
    [#这量的序号对应于dstal中别表的索引
        opts.ParallelAxisOpts(dim=0,name="z_xse"),
        opts.ParallelAxisOpts(dim=1,name="z_num"),
        opts.ParallelAxisOpts(dim=2,name="p_sales"),
        opts.ParallelAxisOpts(dim=3,name="p_bdj"),
        opts.ParallelAxisOpts(dim=4,name="p_price")
    ]
)
c.add("parallel",data,linestyle_opts=opts.LineStyleOpts(width=3,opacity=1))
#c.set_globa1_opts(title_opts=opts.TitleOpts(tit1e="") )
c.render_notebook()
```
